```python
%matplotlib inline  
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = "data/Preprocessed_AFData.xlsx"
df = pd.read_excel(file_path)

# Extract features (first 30 columns) and labels (31st column)
features = df.iloc[:, :30]
labels = df.iloc[:, 30]

# Separate AF and non-AF samples
df_af = features[labels == 1]   # AF cases
df_non_af = features[labels == 0]  # Non-AF cases

# Compute mean distributions
mean_af = df_af.mean()
mean_non_af = df_non_af.mean()
mean_difference = abs(mean_af - mean_non_af)
```

```python
print("cases of af:", len(df_af))
print("cases of non-af:", len(df_non_af))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(mean_af, label="AF (Atrial Fibrillation)", color='red', linewidth=2)
plt.plot(mean_non_af, label="Non-AF (Normal)", color='blue', linewidth=2)
plt.xlabel("R-R Interval Bins (30 bins)")
custom_labels = [f"Bin {i+1}" for i in range(30)]
plt.xticks(ticks=range(30), labels=custom_labels, rotation=90)
plt.ylabel("Normalized Frequency")
plt.title("Comparison of R-R Interval Distributions Between AF and Non-AF")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(custom_labels, mean_difference, color="purple")
plt.xlabel("R-R Interval Bins (30 bins)")
plt.xticks(rotation=90)
plt.ylabel("Absolute Difference in Mean Frequency")
plt.title("Feature Importance: Difference in Mean Distributions (AF vs. Non-AF)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
